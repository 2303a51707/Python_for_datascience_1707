```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temperatures_celsius = np.array([25.3, 26.1, 28.4, 31.2, 33.8, 35.6, 34.2, 33.1, 31.5, 29.4, 27.0, 25.8])
```

```python
average_temp = np.mean(temperatures_celsius)
print(average_temp);
```

```python
hottest_month_index = np.argmax(temperatures_celsius) + 1
coldest_month_index = np.argmin(temperatures_celsius) + 1
```

```python
normalized_temp = (temperatures_celsius - np.min(temperatures_celsius)) / (np.max(temperatures_celsius) - np.min(temperatures_celsius))
print(normalized_temp);
```

```python
months_above_avg = np.where(temperatures_celsius > average_temp)[0] + 1
```

```python
temperatures_fahrenheit = (temperatures_celsius * 9/5) + 32
```

```python
plt.plot(range(1, 13), temperatures_celsius, marker='o', label='°C')
plt.plot(range(1, 13), temperatures_fahrenheit, marker='s', label='°F')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.title('Monthly Temperature Trend')
plt.legend()
plt.grid(True)
plt.show()
```
